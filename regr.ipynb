# Mise en place d'un modèle ML basée sur la régression linéaire 

Le but va être de trouver une fonction d'estimation des données covid. Pour cela nous allons utiliser le début de notre df comme entrainement et la fin comme test. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

df = pd.read_csv("Dataframe_clean.csv",index_col=0)
```

```python
df
```

#### Regression naive

On a 500 valeurs, nous utilerons 2/3 premières valeurs pour le training et le reste pour le test. 

```python
df["x"] = x = np.arange(0,508)
training = df[["conf","x"]][:381]
test = df[["conf","x"]]
print(training)
```

```python

X_training = x[:381,np.newaxis]
X_training = np.insert(X_training,1,[0]*381,axis=1)
print(X_training.size)
X_training[:,1] = 0
print(X_training)
```

```python
import sklearn as skl 
from sklearn.preprocessing import  PolynomialFeatures
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import r2_score


for k in range(1,7):
    poly_training =  PolynomialFeatures(degree=(0,k),include_bias=True).fit_transform(x[:381,np.newaxis])
    poly_test = PolynomialFeatures(degree=(0,k),include_bias=True).fit_transform(x[381:,np.newaxis])
    poly_total = PolynomialFeatures(degree=(0,k),include_bias=True).fit_transform(x[:,np.newaxis])
    model = Ridge(1e-2).fit(poly_training,training["conf"])
    fig,ax = plt.subplots()
    ax.plot(df["conf"], label="Real" )
    ax.plot(model.predict(poly_total), label="predicted")
    tr_score = r2_score(training["conf"],model.predict(poly_training))
    print(f"(degré {k}) score d'entrainement : {tr_score} ")
    te_score = r2_score(test["conf"][381:],model.predict(poly_test))
    print(f"(degré {k}) score de test : {te_score} ")
    total_score = r2_score(test["conf"],model.predict(poly_total))
    print(f"(degré {k}) score total : {total_score} ")
    plt.legend()
    ax.set_title(f"Prediction du nombre de cas pour deg={k}")
    plt.show()



# for k in range(1,6): 
#     poly =  PolynomialFeatures(k).fit_transform(x[:381,np.newaxis])
#     model = Ridge(1e-6).fit(poly,training["conf"])
#     tr_score = r2_score(training,model.predict(training))
#     print(f"(degré {k}) score d'entrainement : {tr_score} ")
#     te_score = r2_score(test,poly.predict(test["x"]))
#     print(f"(degré {k}) score d'entrainement : {te_score} ")
#     fig,ax = plt.subplots()
#     ax.plot(df["conf"], label="Real" )
#     ax.plot(poly.predicts(x[:,np.newaxis]), label="predicted")
```

#### Regression sur de courtes périodes

```python
index_train = [k+i for k in range(508-21) for i in range(14) if k%21 == 0]
train_df = df.loc[index_train]
test_df = df[~df.index.isin(index_train)]
fig, ax = plt.subplots()
ax.scatter(train_df.index,train_df["conf"],marker='.',label="entrainement")
ax.scatter(test_df.index,test_df["conf"],marker='.',color="purple",label="test")
plt.legend()
ax.set_title("Nouvelle répartition d'entrainement/test")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
fig, (ax,ax2) = plt.subplots(2)
d=3
prediction = []
prediction_test = []
score_fin = 1
for k in range(len(df)//21):
    train = df[k*21:k*21+14]
    test = df[k*21+15: (k+1)*21]
    total = df[k*21:(k+1)*21]
    x_train =train["x"]
    x_test = test["x"]
    x_total = total["x"]
    poly_training = PolynomialFeatures(degree=d,include_bias=True).fit_transform(x_train[:,np.newaxis])
    poly_test = PolynomialFeatures(degree=d,include_bias=True).fit_transform(x_test[:,np.newaxis])
    poly_total = PolynomialFeatures(degree=d,include_bias=True).fit_transform(x_total[:,np.newaxis])
    model = Ridge(1e-2).fit(poly_training,train["conf"])
    total_score = r2_score(total["conf"],model.predict(poly_total))
    
    print(f"(degré {k}) score total : {total_score} ")
    ax.plot(total.index,model.predict(poly_total))
    for pred in model.predict(poly_total) : 
        prediction.append(pred)
    for pred in model.predict(poly_test) : 
        prediction_test.append(pred)

#effet de bord 
total = df[(k+1)*21:]
x_total = total["x"]
poly_total = PolynomialFeatures(degree=d,include_bias=True).fit_transform(x_total[:,np.newaxis])
total_score = r2_score(total["conf"],model.predict(poly_total))
print(f"(degré {k}) score total : {total_score} ")
ax.plot(total.index,model.predict(poly_total))
for pred in model.predict(poly_total) : 
    prediction.append(pred)
    prediction_test.append(pred)




ax2.scatter(train_df.index,train_df["conf"],marker='.',label="entrainement")
ax2.scatter(test_df.index,test_df["conf"],marker='.',color="purple",label="test")
ax2.legend()

fig.suptitle("Regression sur de courtes périodes.")
plt.show()


total_score = r2_score(df["conf"],prediction)
print(f"score final : {total_score} ")

# test_total_score = r2_score(test_df["conf"],prediction_test)
# print(f"score final des test : {test_total_score} ")
```
